# Feature Engineering
## Objective

Preparing the dataset for machine learning models by transforming categorical variables, scaling numerical features, and splitting the data into training and testing sets.

```python
import pandas as pd

df = pd.read_csv("../data/preprocessed/churn_clean.csv")
```

```python
X= df.drop('Churn',axis=1)
Y= df['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, Y,
    test_size=0.2,
    stratify=Y,
    random_state=42
)
```

```python
cat_cols = X.select_dtypes(include="object").columns
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
```

```python
X_train = pd.get_dummies(X_train, columns=cat_cols, drop_first=True)
X_test = pd.get_dummies(X_test, columns=cat_cols, drop_first=True)
```

```python
X_train, X_test = X_train.align(X_test, join="left", axis=1, fill_value=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
X_train.to_csv("../data/preprocessed/X_train.csv", index=False)
X_test.to_csv("../data/preprocessed/X_test.csv", index=False)

y_train.to_csv("../data/preprocessed/y_train.csv", index=False)
y_test.to_csv("../data/preprocessed/y_test.csv", index=False)
```
